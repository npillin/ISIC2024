## Description

Over the past several years, dermoscopy-based AI algorithms have been shown to benefit clinicians in diagnosing melanoma, basal cell, and squamous cell carcinoma. Dermatoscope images reveal morphologic features not visible to the naked eye, but these images are typically only captured in dermatology clinics. Algorithms that benefit people in primary care or non-clinical settings must be adept to evaluating lower quality images. This competition leverages 3D TBP to present a novel dataset of every single lesion from thousands of patients across three continents with images resembling cell phone photos.

## Dataset Description
The dataset consists of diagnostically labelled images (SLICE-3D dataset) with additional metadata. The images are JPEGs. The associated .csv file contains a binary diagnostic label (target), potential input variables (e.g. age_approx, sex, anatom_site_general, etc.), and additional attributes (e.g. image source and precise diagnosis).

To mimic non-dermoscopic images, this competition uses standardized cropped lesion-images of lesions from 3D Total Body Photography (TBP). Vectra WB360, a 3D TBP product from Canfield Scientific, captures the complete visible cutaneous surface area in one macro-quality resolution tomographic image. An AI-based software then identifies individual lesions on a given 3D capture. This allows for the image capture and identification of all lesions on a patient, which are exported as individual 15x15 mm field-of-view cropped photos. The dataset contains every lesion from a subset of thousands of patients seen between the years 2015 and 2024 across nine institutions and three continents.

The following are examples from the training set. 'Strongly-labelled tiles' are those whose labels were derived through histopathology assessment. 'Weak-labelled tiles' are those who were not biopsied and were considered 'benign' by a doctor.

- isic_id: Unique case identifier.
- patient_id: Unique patient identifier.
- age_approx: Approximate age of patient at time of imaging.
- sex: Sex of the person.
- anatom_site_general: Location of the lesion on the patient's body.
- clin_size_long_diam_mm: Maximum diameter of the lesion (mm).
- image_type: Structured field of the ISIC Archive for image type.
- tbp_tile_type: Lighting modality of the 3D TBP source image.
- tbp_lv_A: A inside lesion.
- tbp_lv_Aex: A outside lesion.
- tbp_lv_B: B inside lesion.
- tbp_lv_Bext: B outside lesion.
- tbp_lv_C: Chroma inside lesion.
- tbp_lv_Cext: Chroma outside lesion.
- tbp_lv_H: Hue inside the lesion; calculated as the angle of A* and B* in LAB* color space. Typical values range from 25 (red) to 75 (brown).
- tbp_lv_Hext: Hue outside lesion.
- tbp_lv_L: L inside lesion.
- tbp_lv_Lext: L outside lesion.
- tbp_lv_areaMM2: Area of lesion (mm^2).
- tbp_lv_area_perim_ratio: Border jaggedness, the ratio between lesions perimeter and area. Circular lesions will have low values; irregular shaped lesions will have higher values. Values range 0-10.
- tbp_lv_color_std_mean: Color irregularity, calculated as the variance of colors within the lesion's boundary.
- tbp_lv_deltaA: Average A contrast (inside vs. outside lesion).
- tbp_lv_deltaB: Average B contrast (inside vs. outside lesion).
- tbp_lv_deltaL: Average L contrast (inside vs. outside lesion).
- tbp_lv_deltaLBnorm: Contrast between the lesion and its immediate surrounding skin. Low contrast lesions tend to be faintly visible such as freckles; high contrast lesions tend to be those with darker pigment. Calculated as the average delta LB of the lesion relative to its immediate background in LAB* color space. Typical values range from 5.5 to 25.
- tbp_lv_eccentricity: Eccentricity.
- tbp_lv_location: Classification of anatomical location, divides arms & legs to upper & lower; torso into thirds.
- tbp_lv_location_simple: Classification of anatomical location, simple.
- tbp_lv_minorAxisMM: Smallest lesion diameter (mm).
- tbp_lv_nevi_confidence: Nevus confidence score (0-100 scale) is a convolutional neural network classifier estimated probability that the lesion is a nevus. The neural network was trained on approximately 57,000 lesions that were classified and labeled by a dermatologist.
- tbp_lv_norm_border: Border irregularity (0-10 scale); the normalized average of border jaggedness and asymmetry.
- tbp_lv_norm_color: Color variation (0-10 scale); the normalized average of color asymmetry and color irregularity.
- tbp_lv_perimeterMM: Perimeter of lesion (mm).
- tbp_lv_radial_color_std_max: Color asymmetry, a measure of asymmetry of the spatial distribution of color within the lesion. This score is calculated by looking at the average standard deviation in LAB* color space within concentric rings originating from the lesion center. Values range 0-10.
- tbp_lv_stdL: Standard deviation of L inside lesion.
- tbp_lv_stdLExt: Standard deviation of L outside lesion.
- tbp_lv_symm_2axis: Border asymmetry; a measure of asymmetry of the lesion's contour about an axis perpendicular to the lesion's most symmetric axis. Lesions with two axes of symmetry will therefore have low scores (more symmetric), while lesions with only one or zero axes of symmetry will have higher scores (less symmetric). This score is calculated by comparing opposite halves of the lesion contour over many degrees of rotation. The angle where the halves are most similar identifies the principal axis of symmetry, while the second axis of symmetry is perpendicular to the principal axis. Border asymmetry is reported as the asymmetry value about this second axis. Values range 0-10.
- tbp_lv_symm_2axis_angle: Lesion border asymmetry angle.
- tbp_lv_x: X-coordinate of the lesion on 3D TBP.
- tbp_lv_y: Y-coordinate of the lesion on 3D TBP.
- tbp_lv_z: Z-coordinate of the lesion on 3D TBP.
- attribution: Image attribution, synonymous with image source.
- copyright_license: Copyright license.

## Data Preprocessing

```python
# These columns won't be used for model training and predictions
# We'll keep isic_id as it will be useful later on to prepare the data loaders
cols_to_drop = ['patient_id','image_type','copyright_license','attribution']

# Create pipeline
set_config(transform_output='pandas') # Set output format to pandas df

# Define imputer ColumnTransformer
imputer = ColumnTransformer(
    [
        ('num_imputer', SimpleImputer(strategy='mean'), num_cols),
        ('cat_imputer', SimpleImputer(strategy='most_frequent'), cat_cols)
    ],
    remainder='passthrough',
    verbose_feature_names_out=False # Don't add prefix to column names
)

# Define encoder ColumnTransformer
encoder = ColumnTransformer(
    [('ohe', OneHotEncoder(drop='first', sparse_output=False), cat_cols)],
    remainder='passthrough',
    verbose_feature_names_out=False # Don't add prefix to column names
)

# Define sclaer ColumnTransformer
scaler = ColumnTransformer(
    [('std_scaler', StandardScaler(), make_column_selector(dtype_include=np.number))], # Don't apply on isic_id
    remainder='passthrough',
    verbose_feature_names_out=False # Don't add prefix to column names
)

# Define pipeline
pipe = Pipeline([
    # Drop columns that won't be used for training or predictions
    ('cols_dropper', FunctionTransformer(lambda df: df.drop(cols_to_drop, axis=1))),
    # Handle missing values
    ('imputer', imputer),
    # Handle categorical columns
    ('encoder', encoder),
    # Scale data
    ('scaler', scaler)
])
```

```python
# Define classes that will help for data loading during training, evaluation and testing
# Positive cases are very rare and we'll use oversample them and use data augmentation
# We'll keep the same number of nugative cases, randomly picked

class TrainDataset:
    """
        df: Original metadata
        df_pp: Preprocessed metadata
        data_h5: Path to images in HDF5 format
        transform: Transformations to be applied to images
        oversampling: Number of times the positive cases are to be duplicated
        train_ratio: Fraction of train data to be used for training, the rest is for validation
    """
    def __init__(self, df, df_pp, data_h5, transform=None, oversampling=3, train_ratio=0.9):
        
        # Create train and val data
        df_positive = df.loc[df['target']==1, ['isic_id','target']].reset_index(drop=True) # Select all positive cases
        df_train_positive = df_positive[:int(len(df_positive) * train_ratio)] # Train/val split
        df_val_positive = df_positive[int(len(df_positive) * train_ratio):]
        
        df_train_positive_ovr = df_train_positive.copy()
        df_val_positive_ovr = df_val_positive.copy()
        for _ in range(oversampling): # Oversampling of positive cases
            df_train_positive_ovr = pd.concat([df_train_positive_ovr, df_train_positive]).reset_index(drop=True)
            df_val_positive_ovr = pd.concat([df_val_positive_ovr, df_val_positive]).reset_index(drop=True)
        
        df_negative = df.loc[df['target']==0, ['isic_id','target']].sample(frac=1).reset_index(drop=True) # Select all negative cases and shuffle them
        df_negative = df_negative[:len(df_train_positive_ovr) + len(df_val_positive_ovr)] # Keep enough negative cases to cover train and val positive cases
        df_train_negative = df_negative[:len(df_train_positive_ovr)] # Save negative cases for train/val
        df_val_negative = df_negative[len(df_train_positive_ovr):len(df_train_positive_ovr) + len(df_val_positive_ovr)]
        
        # Save train and val dataframes
        self.df_train = pd.concat([df_train_positive_ovr, df_train_negative]).sample(frac=1).reset_index(drop=True)
        self.df_val = pd.concat([df_val_positive_ovr, df_val_negative]).sample(frac=1).reset_index(drop=True)
        
        # Save relevant preprocessed metadata for train and val and match indexes
        self.df_train_pp = self.df_train[['isic_id']].merge(df_pp, on='isic_id', how='left').drop('isic_id', axis=1).reset_index(drop=True)
        self.df_val_pp = self.df_val[['isic_id']].merge(df_pp, on='isic_id', how='left').drop('isic_id', axis=1).reset_index(drop=True)
        
        # Save path to HDF5 file and image transfomations
        self.h5 = h5py.File(data_h5, 'r')
        self.transform = transform
        
    def __len__(self):
        return len(self.df_train)

    def __getitem__(self, idx):
        key = self.df_train.loc[idx, 'isic_id']
        img = Image.open(BytesIO(self.h5[key][()]))
        if self.transform:
            img = self.transform(img)
        meta = torch.FloatTensor(self.df_train_pp.loc[idx].to_numpy()).view(-1) # Recover preprocessed metadata
        target = self.df_train.loc[idx, 'target']
        return img, meta, np.float32(target)

        
class ValDataset:
    """
        A simplified version of TrainDataset that recovers the validation data
        from train_dataset and makes it available.
    """
    def __init__(self, train_dataset, transform=None):
        self.df_val = train_dataset.df_val
        self.df_val_pp = train_dataset.df_val_pp
        
        self.h5 = train_dataset.h5
        self.transform = train_dataset.transform
        
    def __len__(self):
        return len(self.df_val)
        
    def __getitem__(self, idx):
        key = self.df_val.loc[idx, 'isic_id']
        img = Image.open(BytesIO(self.h5[key][()]))
        if self.transform:
            img = self.transform(img)
        meta = torch.FloatTensor(self.df_val_pp.iloc[idx].to_numpy()).view(-1) # Recover metadata
        target = self.df_val.loc[idx, 'target']
        return img, meta, np.float32(target)

        
class TestDataset:
    def __init__(self, df, df_pp, data_h5, transform=None):
        self.df_pp = df_pp
        self.h5 = h5py.File(data_h5, 'r')
        self.keys = list(self.h5.keys())
        
        self.transform = transform
        
    def __len__(self):
        return len(self.keys)

    def __getitem__(self, idx):
        key = self.keys[idx]
        img = Image.open(BytesIO(self.h5[key][()]))
        if self.transform:
            img = self.transform(img)
        meta = torch.FloatTensor(self.df_pp.iloc[idx].to_numpy()).view(-1) # Recover metadata
        return img, meta
```

```python
# Define baseic image transformer for validation and making final predictions
img_size = 128 # 12x128 images
basic_transform = transforms.Compose(
    [
        transforms.Resize((img_size, img_size)),  # Images have varying sizes : resize to 128x128
        transforms.ToTensor(),  # Convert image to PyTorch tensor
        transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))  # Normalize image
    ]
)

# Define image transformer for training, with data augmentation
train_transform = transforms.Compose(
    [
        transforms.Resize((img_size, img_size)),
        transforms.ToTensor(),
        transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)),
        transforms.RandomHorizontalFlip(),
        transforms.RandomVerticalFlip(),
        transforms.RandomAffine( # Randomly rotate, translate and scale image
            degrees=(-10,10),
            translate=(0.1,0.1),
            scale=(0.9,1.1)
        ), 
    ]
)

# Define train data loader
train_dataset = TrainDataset(
    df=train_metadata,
    df_pp=X_train,
    data_h5='/kaggle/input/isic-2024-challenge/train-image.hdf5',
    transform=train_transform,
    oversampling=4,
)
print(f"Number of train data points: {len(train_dataset)}")
train_data_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)


# Define val data loader
val_dataset = ValDataset(train_dataset, transform=basic_transform)
print('-'*50)
print(f"Number of val data points: {len(val_dataset)}")
val_data_loader = DataLoader(val_dataset, batch_size=32, shuffle=True)
```

## Modeling

```python
# Define CNN
# This is a hybrid model: the metadata is injected in one of the last
# layers, along the features extracted by the CNN from the image

class hybrid_model(nn.Module):
    def __init__(self):
        super(hybrid_model, self).__init__()
        
        self.CNN = nn.Sequential(
            
            nn.Conv2d(3, 64, kernel_size=3), # 3x128x128 -> 64x126x126
            nn.ReLU(),
            nn.MaxPool2d(2), # 64x126x126 -> 64x63x63
            
            nn.Conv2d(64, 32, kernel_size=3), # 64x63x63 -> 32x61x61
            #nn.Dropout2d(),
            nn.ReLU(),
            nn.MaxPool2d(2), # 32x61x61 -> 32x30x30
            
            nn.Conv2d(32, 32, kernel_size=3), # 32x30x30 -> 32x28x28
            #nn.Dropout2d(),
            nn.ReLU(),
            nn.MaxPool2d(2), # 32x28x28 -> 32x14x14
            
            nn.Conv2d(32, 16, kernel_size=3), # 32x14x14 -> 16x12x12
            #nn.Dropout2d(),
            nn.ReLU(),
            nn.MaxPool2d(2), # 16x12x12 -> 16x6x6
            
            nn.Flatten(), # 16x6x6 -> 576
            nn.Linear(576, 100),
            nn.BatchNorm1d(100),
            nn.ReLU()
        )
        
        self.hybrid_classifier = nn.Sequential(
            nn.Linear(100+38, 128), # 100 CNN output features + 38 preprocessed metadata columns
            nn.BatchNorm1d(128),
            nn.SiLU(),
            nn.Dropout(),
            nn.Linear(128, 64),
            nn.BatchNorm1d(64),
            nn.SiLU(),
            nn.Dropout(),
            nn.Linear(64, 32),
            nn.BatchNorm1d(32),
            nn.SiLU(),
            nn.Dropout(),
            nn.Linear(32, 1),
            nn.Sigmoid()
        )

    def forward(self, img, meta):
    #def forward(self, meta):
        X = self.CNN(img) # Process image through CNN
        X = self.hybrid_classifier(torch.cat((X, meta), dim=1)) # Combine CNN output feature with preprocessed metadata
        #X = self.hybrid_classifier(meta) # Combine CNN output feature with preprocessed metadata
        return X

model = hybrid_model()
```

```python
print(torch.cuda.get_device_name(0))
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
model.to(device)
```

```python
# Define loss function and optimizer
criterion = nn.BCELoss()
optimizer = optim.AdamW(model.parameters(), lr=0.01, weight_decay=1e-4)
```

```python
# Helper function to evaluate model's performance during training, using validation data
def eval_model(model, criterion, dataloader):
    model.eval()
    total_loss = 0
    total_correct = 0
    total_samples = 0
    val_targets = torch.empty(0, dtype=torch.float)
    val_outputs = torch.empty(0, dtype=torch.float)
    with torch.no_grad():
        for i, (images, metas, targets) in enumerate(dataloader):
            val_targets = torch.cat((val_targets, targets))
            
            images = images.to(device)
            metas = metas.to(device)
            targets = targets.to(device)
            
            outputs = model(img=images, meta=metas)
            #outputs = model(meta=metas) ###
            
            loss = criterion(outputs, targets.unsqueeze(1))
            
            total_loss += loss.item()
            total_correct += ((outputs > 0.5) == targets).sum().item()
            total_samples += targets.size(0)
            
            val_outputs = torch.cat((val_outputs, outputs.cpu()))
    
    return (
        total_loss / len(dataloader),
        total_correct / total_samples,
        val_targets,
        val_outputs
    )
```

```python
# Training hyperparameters and global variables
lr_patience = 5
patience = 10

epoch = 0

train_losses = []
val_losses = []

best_loss = np.inf
ind_best_loss = 0
best_model_wts = model.state_dict()
last_change = 0
```

```python
# Train model
nb_epoch = 30 # Max number of epochs to iterate at a time
epoch_start = epoch

for ep in range(nb_epoch):
    model.train()
    running_loss = 0.0
    for i, (images, metas, labels) in enumerate(train_data_loader):
        images, metas, labels = images.to(device), metas.to(device), labels.to(device).reshape(-1, 1)

        # Training step
        optimizer.zero_grad()
        outputs = model(img=images, meta=metas)
        #outputs = model(meta=metas) ###
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
    
    # Update epoch counter
    epoch += 1
    
    # Compute and save losses
    train_loss = running_loss / len(train_data_loader)
    train_losses.append(train_loss)
    val_loss, _, _, _ = eval_model(model, criterion, val_data_loader)
    val_losses.append(val_loss)
    
    # Track best loss and save model
    if(val_loss < best_loss):
        best_loss = val_loss
        ind_best_loss = epoch
        best_model_wts = model.state_dict()
        
    # Lower learning rate if val loss doesn't improve
    if(epoch-max(ind_best_loss, last_change) >= lr_patience):
        last_change = epoch
        for param_group in optimizer.param_groups:
            param_group['lr'] *= 0.1
        
    if(epoch-ind_best_loss > patience):
        print(f'Early stopping on epoch {epoch}')
        break
    
    print(f'Epoch [{epoch}/{epoch_start+nb_epoch}], train_loss: {train_loss:.4f}, val_loss: {val_loss:.4f}, lr: {optimizer.param_groups[0]["lr"]:.5f}, best_loss: {best_loss:.4f}, at step {ind_best_loss}, {epoch-ind_best_loss} steps ago')
    
model.load_state_dict(best_model_wts)
```
